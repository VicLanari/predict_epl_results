```python
import pandas as pd
```

```python
LINKS = ["mmz4281/2223/E0.csv",
 "mmz4281/2122/E0.csv",
 "mmz4281/2021/E0.csv",
 "mmz4281/1920/E0.csv",
 "mmz4281/1819/E0.csv",
 "mmz4281/1718/E0.csv",
 "mmz4281/1617/E0.csv",
 "mmz4281/1516/E0.csv",
 "mmz4281/1415/E0.csv",
 "mmz4281/1314/E0.csv",
 "mmz4281/1213/E0.csv",
 "mmz4281/1112/E0.csv",
 "mmz4281/1011/E0.csv",
 "mmz4281/0910/E0.csv",
 "mmz4281/0809/E0.csv",
 "mmz4281/0708/E0.csv",
 "mmz4281/0607/E0.csv",
 "mmz4281/0506/E0.csv"]


def grab_epl_data():
    df = pd.DataFrame()
    for link in LINKS:
        df = pd.concat([df, pd.read_csv(f"https://www.football-data.co.uk/{link}")]).reset_index(drop=True)
    return df
```

```python
def get_odds_final(df):
    df_odds = df[["B365H", "B365D", "B365A", "BWH", "BWD", "BWA", "IWH", "IWD",\
    "IWA", "PSH", "PSD", "PSA", "WHH", "WHD", "WHA", "VCH", "VCD", "VCA",]]
    df_odds["AHO"] = (df_odds["B365H"] + df_odds["BWH"] + df_odds["IWH"] + df_odds["WHH"] + df_odds["VCH"])/5
    df_odds["AAO"] = (df_odds["B365A"] + df_odds["BWA"] + df_odds["IWA"] + df_odds["WHA"] + df_odds["VCA"])/5
    print(df_odds.shape, df_odds.shape)
    df_odds = df_odds[["AAO", "AHO"]]
    print(df_odds.shape)
    return df_odds
```

```python
def stats_features(df):
    stat_cols = ['AC', 'AF', 'AR', 'AS', 'AST', 'AY', 'HC', 'HF', 'HR', 'HS', 'HST', 'HY', 'margin']
    df['margin'] = df['FTHG'] - df['FTAG']
    df_stats = df[stat_cols]
    # df_stats = df_stats.dropna()
    df_stats['THC'] = df_stats['HY'] + df_stats['HR']
    df_stats['TAC'] = df_stats['AY'] + df_stats['AR']
    final_cols = ['AST', 'HST', 'HC', 'AC', 'THC', 'TAC', 'margin']
    df_final = df_stats[final_cols]
    
    return df_final     
```

```python
df = grab_epl_data()
```

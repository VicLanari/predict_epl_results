y_compare['correct'] = 0
y_compare.head()

```python
y_compare.columns.values
```

```python
y_compare = y_compare.rename(columns={0:'prediction'})
y_compare.head()
```

```python
for row in pd.DataFrame.iterrows(y_compare):
    # print(row[1]['margin'])
    if row[1]['margin'] == 0:
        if row[1]['prediction'] < 1 and row[1]['prediction'] > -1:
            row[1]['correct'] = 1
    if row[1]['margin'] > 0:
        if row[1]['prediction'] > 0:
            row[1]['correct'] = 1
    if row[1]['margin'] < 0:
        if row[1]['prediction'] < 0:
            row[1]['correct'] = 1
```

```python
def check_conditions(margin, prediction):
    output = 0
    if margin == 0:
        if prediction < 1 and prediction > -1:
            output = 1
    if margin > 0:
        if prediction > 0:
            output = 1
    if margin < 0:
        if prediction < 0:
            output = 1
    return output
```

```python
y_compare['correct'] = y_compare.apply(lambda x: check_conditions(x['margin'], x['prediction']), axis=1)
```

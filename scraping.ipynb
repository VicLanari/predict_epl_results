```python
def grab_epl_data():
    df = pd.DataFrame()
    for link in links:
        # print(f"https://www.football-data.co.uk/{link}")
        df = pd.concat([df, pd.read_csv(f"https://www.football-data.co.uk/{link}")])
    return df
```
